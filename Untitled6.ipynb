```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data: blood sugar levels of 200 patients
# (normally distributed around 110 with some variation)
blood_sugar = np.random.normal(loc=110, scale=25, size=200)

# Create the density plot
sns.kdeplot(blood_sugar, fill=True, color='Black')

# Add titles and labels
plt.title("Density Plot of Blood Sugar Levels")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")

# Show the plot
plt.show()
```

```python
# Install Plotly if you haven't
# pip install plotly pandas

import pandas as pd
import plotly.express as px

# -------------------------
# Step 1: Create the dataset
# -------------------------
data = {
    'State': ['Maharashtra', 'Kerala', 'Delhi', 'Karnataka', 'Tamil Nadu'],
    'Cases': [320, 250, 150, 180, 200]
}

df = pd.DataFrame(data)

# -------------------------
# Step 2: Create the map
# -------------------------
# Plotly uses location names and matches them automatically for India
fig = px.choropleth(
    df,
    locations='State',          # column containing state names
    locationmode='geojson-id',  # automatically detects states by name
    color='Cases',              # column to color by
    hover_name='State',         # hover label
    color_continuous_scale='Reds',  # red shades indicate higher cases
    title='COVID-19 Cases Across Indian States (2020)'
)

# -------------------------
# Step 3: Adjust map focus
# -------------------------
fig.update_geos(
    fitbounds="locations",
    visible=False,
    projection_type="mercator"
)

fig.show()
```
